```python
from graphviz import Graph
from operator import eq
from itertools import combinations_with_replacement, combinations, permutations, product
```

```python
def getBinStrings(n):
    return list(product(range(2), repeat=n))
```

```python
def getNCubeEdges(n):
    return [pair for pair in combinations(getBinStrings(n), 2)
                          if sum(eq(*x) for x in (zip(*pair))) == n-1]
```

```python
def tupToStr(tup):
    return "".join(str(x) for x in tup)
def tupToStrLst(tupList):
    return [tupToStr(x) for x in tupList]
```

```python
def getNCubeGraph(n):
    nodes = tupToStrLst(getBinStrings(n))
    edges = [tupToStrLst(x) for x in getNCubeEdges(n)]
    
    cube = Graph()
    for x in nodes:
        cube.node(x, x)
        
    for x in edges:
        cube.edge(*x)
        
    return cube
```
